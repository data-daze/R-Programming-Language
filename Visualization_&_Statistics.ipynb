# ggplot2: Grammar of Graphics

`ggplot2` is based on the **Grammar of Graphics** idea:

1. **Data** → Which dataset are you using?  
   Example: `mtcars`  

2. **Aesthetics (aes)** → How variables map to visuals (X, Y, color, size, shape).  
   Example: `aes(x = wt, y = mpg)`  

3. **Geoms** → What kind of plot do you want?  
   - `geom_point()` → scatter plot  
   - `geom_bar()` → bar plot  
   - `geom_histogram()` → histogram  

4. **Layers** → You keep adding details with `+`.  
   - `+ labs()` → labels (title, axis names)  
   - `+ theme()` → style (fonts, background)  
   - `+ g


````r
# ggplot2 Basics

# ggplot2 is more powerful, flexible, and uses the Grammar of Graphics.

# First, load the library

library(ggplot2)


#We’ll use the built-in dataset mtcars.
````

````r
ggplot(mtcars, aes(x = wt, y = mpg)) +
  geom_point(color = "blue", size = 3) +
  labs(title = "Scatter Plot with ggplot2", x = "Weight", y = "Miles per Gallon")
````

### ggplot2 Example with `mtcars` Dataset

- `ggplot(mtcars, ...)` → Starts a plot using the **mtcars** dataset.  
- `aes(x = wt, y = mpg)` → Aesthetic mappings (connects variables to visuals).  
  - `x = wt` → Car weight goes on the **X-axis**.  
  - `y = mpg` → Miles per gallon goes on the **Y-axis**.  

- `+ geom_point(color = "blue", size = 3)` → Adds a scatter plot layer.  
  - `geom_point()` → Draws points.  
  - `color = "blue"` → All points are blue.  
  - `size = 3` → Points are larger (size 3).  

- `+ labs(title = ..., x = ..., y = ...)` → Adds labels.  
  - `title = "Scatter Plot with ggplot2"` → Plot title.  
  - `x = "Weight"` → X-axis label.  
  - `y = "Miles per Gallon"` → Y-axis label.  


````r
#Bar plot
ggplot(mtcars, aes(x = factor(cyl))) +
  geom_bar(fill = "lightblue", color = "black") +
  labs(title = "Bar Plot of Cylinders", x = "Cylinders", y = "Count")
````

### ggplot2 Bar Plot Example with `mtcars` Dataset

- `ggplot(mtcars, aes(x = factor(cyl)))` → Starts a plot using the **mtcars** dataset.  
  - `aes(x = factor(cyl))` → Aesthetic mapping (connects variable to visuals).  
    - `x = factor(cyl)` → Number of cylinders on the **X-axis** as a **categorical variable**.  

- `+ geom_bar(fill = "lightblue", color = "black")` → Adds a bar plot layer.  
  - `geom_bar()` → Draws bars counting the number of cars per cylinder group.  
  - `fill = "lightblue"` → Fills the bars with light blue color.  
  - `color = "black"` → Adds a black border to the bars.  

- `+ labs(title = ..., x = ..., y = ...)` → Adds labels.  
  - `title = "Bar Plot of Cylinders"` → Plot title.  
  - `x = "Cylinders"` → X-axis label.  
  - `y = "Count"` → Y-axis label (ggplot2 automatically counts the number of cars in each cylinder group).  


````r
#histogram
ggplot(mtcars, aes(x = mpg)) +
  geom_histogram(binwidth = 2, fill = "lightgreen", color = "black") +
  labs(title = "Histogram of MPG", x = "Miles per Gallon", y = "Frequency")
````

### ggplot2 Histogram Example with `mtcars` Dataset

- `ggplot(mtcars, aes(x = mpg))` → Starts a plot using the **mtcars** dataset.  
  - `aes(x = mpg)` → Aesthetic mapping (connects variable to visuals).  
    - `x = mpg` → Miles per gallon on the **X-axis**.  

- `+ geom_histogram(binwidth = 2, fill = "lightgreen", color = "black")` → Adds a histogram layer.  
  - `geom_histogram()` → Draws a histogram.  
  - `binwidth = 2` → Groups data into bins of width 2 (e.g., 10–12, 12–14, etc.).  
  - `fill = "lightgreen"` → Fills the bars with green color.  
  - `color = "black"` → Adds black borders to the bars.  

- `+ labs(title = ..., x = ..., y = ...)` → Adds labels.  
  - `title = "Histogram of MPG"` → Plot title.  
  - `x = "Miles per Gallon"` → X-axis label.  
  - `y = "Frequency"` → Y-axis label (counts how many cars fall into each bin).  


### Important Rules of ggplot2

- **Always start with `ggplot(data, aes(...))`**  
  - This initializes the plot and specifies the dataset and aesthetic mappings.  

- **Use `+` to add layers**  
  - Layers can include geoms (points, bars, lines), labels, themes, and more.  

- **`aes()` defines mapping from variables → visuals**  
  - Maps data columns to axes, color, size, shape, etc.  
  - Example: `aes(x = wt, y = mpg, color = cyl)`  

- **Inside geoms, styles can be constants or data-driven**  
  - **Constant style (applied to all points/bars)**:  
    - Example: `geom_point(color = "red")` → all points red  
  - **Data-driven style (varies with data)**:  
    - Example: `geom_point(aes(color = cyl))` → color depends on `cyl` value

### Probability Distributions

#### 1. Probability
- Probability measures how likely an event is to happen.  
- Ranges from 0 (impossible) to 1 (certain).  
- **Example:** Tossing a fair coin → Probability of heads = 0.5  

#### 2. Probability Distributions
- A probability distribution describes how probabilities are distributed over the values of a random variable.  
- **Types:**
  - **Normal distribution**: Continuous, bell-shaped. Defined by mean (μ) and standard deviation (σ).  
  - **Binomial distribution**: Discrete, counts the number of successes in a fixed number of trials. Defined by number of trials (n) and success probability (p).  

#### Key Functions in R

| Function | Description |
|----------|-------------|
| `dnorm(x, mean, sd)` | Probability density at `x` (normal distribution) |
| `rnorm(n, mean, sd)` | Generate `n` random numbers (normal distribution) |
| `pnorm(q, mean, sd)` | Cumulative probability `P(X ≤ q)` (normal distribution) |
| `dbinom(x, size, prob)` | Probability of `x` successes (binomial distribution) |
| `rbinom(n, size, prob)` | Generate `n` random binomial outcomes |


### Normal Distribution

- **`dnorm(x, mean, sd)`** → Probability density at `x` (height of the probability density function).  
- **`rnorm(n, mean, sd)`** → Generate `n` random numbers from a normal distribution.  
- **`pnorm(q, mean, sd)`** → Cumulative probability `P(X ≤ q)` (area under the curve up to `q`).  

**Example:**

```r
# Normal distribution examples
dnorm(0, mean = 0, sd = 1)   # Density at x = 0
rnorm(5, mean = 0, sd = 1)   # Generate 5 random numbers
pnorm(1, mean = 0, sd = 1)   # Cumulative probability P(X ≤ 1)


````r
# Generate 1000 random normal numbers
set.seed(123)
x <- rnorm(1000, mean = 50, sd = 10)

# Density at x = 60
dnorm(60, mean = 50, sd = 10)

# Cumulative probability P(X <= 60)
pnorm(60, mean = 50, sd = 10)

# Histogram
hist(x, breaks = 30, col = "lightblue", main = "Normal Distribution",
     xlab = "Value")
````

### Binomial Distribution

- **`dbinom(x, size, prob)`** → Probability of exactly `x` successes in `size` trials with success probability `prob`.  
- **`rbinom(n, size, prob)`** → Generate `n` random numbers from a binomial distribution.  

**Example:**

```r
# Binomial distribution examples
dbinom(3, size = 5, prob = 0.5)   # Probability of 3 successes in 5 trials
rbinom(5, size = 5, prob = 0.5)   # Generate 5 random binomial outcomes


````r
# Probability of 3 successes in 5 trials, p=0.6
dbinom(3, size = 5, prob = 0.6)

# Generate 100 random binomial outcomes
binom_samples <- rbinom(100, size = 5, prob = 0.6)
table(binom_samples)

# Bar plot of outcomes
barplot(table(binom_samples), col = "pink", main = "Binomial Distribution")
````

### Hypothesis Testing

#### 1. Hypothesis
- A hypothesis is a statement about a population parameter that we want to test.

#### 2. Types
- **Null hypothesis (H₀):** No effect or difference (e.g., "nothing is happening").  
- **Alternative hypothesis (H₁):** There is an effect or difference (e.g., "something is happening").  

#### 3. Common Tests

| Test | Purpose |
|------|---------|
| t-test | Compare means (one-sample, two-sample, paired) |
| Chi-square test | Test association between categorical variables |

#### 4. Interpretation
- **p-value < 0.05** → Reject H₀ → Significant difference/association  
- **p-value ≥ 0.05** → Fail to reject H₀ → No significant difference/association  

---

### Easy Examples

#### Example 1: One-sample t-test
- Suppose average student score = 50 (H₀).  
- Collected scores: 55, 60, 62, 58, 65, 70  
- **Question:** Are these scores significantly different from 50?

#### Example 2: Two-sample t-test
- Compare two classes:  
  - Class A scores: 5, 7, 8, 6, 9  
  - Class B scores: 10, 12, 11, 9, 13  
- **Question:** Are class scores different on average?

#### Example 3: Paired t-test
- Measure before & after a training program:  
  - Before: 100, 102, 98, 95, 105  
  - After: 98, 101, 97, 94, 103  
- **Question:** Did the program change scores?

#### Example 4: Chi-square test
- Survey 70 people:  
  - Male: 20 Yes, 15 No  
  - Female: 30 Yes, 35 No  
- **Question:** Are “Yes/No” responses independent of gender?


````r
# One-sample t-test
scores <- c(55, 60, 62, 58, 65, 70)
t.test(scores, mu = 50)
````

````r
# Two-sample t-test
group1 <- c(5, 7, 8, 6, 9)
group2 <- c(10, 12, 11, 9, 13)
t.test(group1, group2)
````

````r


# Paired t-test
before <- c(100, 102, 98, 95, 105)
after  <- c(98, 101, 97, 94, 103)
t.test(before, after, paired = TRUE)
````

````r




# Chi-square test
data <- matrix(c(20, 15, 30, 35), nrow = 2)
colnames(data) <- c("Yes", "No")
rownames(data) <- c("Male", "Female")
chisq.test(data)
````

### Key Points for Easy Understanding

- **Normal Distribution**  
  - Continuous, bell-shaped  
  - Mean (μ) and Standard Deviation (σ) are important  

- **Binomial Distribution**  
  - Discrete, counts number of successes  
  - Number of trials (n) and success probability (p) are important  

- **t-test** → Compare means (one-sample, two-sample, paired)  

- **Chi-square test** → Check association between categorical variables  

- **p-value Interpretation:**  
  - `p-value < 0.05` → Significant → Reject H₀  
  - `p-value ≥ 0.05` → Not significant → Fail to reject H₀


````r
# Load libraries
library(ggplot2)
library(gridExtra)

# -----------------------------
# 1. Normal Distribution
# -----------------------------
set.seed(123)
normal_data <- rnorm(1000, mean = 50, sd = 10)

p1 <- ggplot(data.frame(x = normal_data), aes(x = x)) +
  geom_histogram(aes(y = ..density..), bins = 30, fill = "lightblue", color = "black") +
  geom_density(color = "red", size = 1) +
  labs(title = "Normal Distribution", x = "Score", y = "Density") +
  theme_minimal()

# -----------------------------
# 2. Binomial Distribution
# -----------------------------
binom_data <- rbinom(1000, size = 5, prob = 0.6)

p2 <- ggplot(data.frame(x = binom_data), aes(x = factor(x))) +
  geom_bar(fill = "pink", color = "black") +
  labs(title = "Binomial Distribution", x = "Number of Heads", y = "Frequency") +
  theme_minimal()

# -----------------------------
# 3. One-sample t-test
# -----------------------------
scores <- c(55, 60, 62, 58, 65, 70)
t_test_result <- t.test(scores, mu = 50)

p3 <- ggplot(data.frame(x = scores), aes(x = x, y = 0)) +
  geom_point(size = 4, color = "blue") +
  geom_vline(xintercept = 50, linetype = "dashed", color = "red", size = 1) +
  labs(title = paste("One-sample t-test\np =", round(t_test_result$p.value,4)), x = "Score", y = "") +
  theme_minimal() +
  theme(axis.text.y = element_blank(), axis.ticks.y = element_blank())

# -----------------------------
# 4. Two-sample t-test
# -----------------------------
group1 <- c(5, 7, 8, 6, 9)
group2 <- c(10, 12, 11, 9, 13)
t_test_2 <- t.test(group1, group2)

df_two <- data.frame(
  score = c(group1, group2),
  group = factor(c(rep("Group 1", length(group1)), rep("Group 2", length(group2))))
)

p4 <- ggplot(df_two, aes(x = group, y = score, fill = group)) +
  geom_boxplot() +
  geom_jitter(width = 0.1, color = "black") +
  labs(title = paste("Two-sample t-test\np =", round(t_test_2$p.value,4)), x = "Group", y = "Score") +
  theme_minimal()

# -----------------------------
# 5. Chi-square Test
# -----------------------------
chi_data <- matrix(c(20, 15, 30, 35), nrow = 2)
colnames(chi_data) <- c("Yes", "No")
rownames(chi_data) <- c("Male", "Female")
chi_test <- chisq.test(chi_data)

df_chi <- as.data.frame(as.table(chi_data))
colnames(df_chi) <- c("Gender", "Response", "Count")

p5 <- ggplot(df_chi, aes(x = Response, y = Count, fill = Gender)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = paste("Chi-square Test\np =", round(chi_test$p.value,4)), x = "Response", y = "Count") +
  theme_minimal()

# -----------------------------
# Combine all plots in one page
# -----------------------------
grid.arrange(p1, p2, p3, p4, p5, ncol = 2)
````

### Correlation

#### Concept
- Correlation measures how two variables move together.  
- **Range:** -1 to 1  
  - `+1` → Perfect positive correlation (both increase together)  
  - `0` → No correlation  
  - `-1` → Perfect negative correlation (one increases, other decreases)  

#### Examples
- Height vs. weight → Usually positively correlated  
- Hours spent watching TV vs. exam scores → Possibly negatively correlated


````r
cor(x, y)  # Pearson correlation by default
````

````r
#Correlation
# Sample data
height <- c(150, 160, 170, 180, 190)
weight <- c(50, 60, 65, 80, 90)

# Calculate correlation
cor(height, weight)  # Positive correlation

# Visualize correlation
plot(height, weight, main="Height vs Weight",
     xlab="Height (cm)", ylab="Weight (kg)",
     pch=19, col="blue")
abline(lm(weight ~ height), col="red")  # Add regression line
````

### Linear Regression

#### Concept
- Linear regression models the relationship between a **dependent variable (Y)** and an **independent variable (X)**.  

#### Equation
\[
Y = \beta_0 + \beta_1 X + \epsilon
\]

- **Y** → Dependent variable  
- **X** → Independent variable  
- **β₀** → Intercept (value of Y when X = 0)  
- **β₁** → Slope (change in Y for a unit change in X)  
- **ϵ** → Error term  

#### Use Cases
- Predicting house price from size  
- Predicting sales from advertising


````r
model <- lm(Y ~ X, data = df)
summary(model)
````

````r
#Linear Regression
# Create data frame
df <- data.frame(height = height, weight = weight)

# Fit linear regression model
model <- lm(weight ~ height, data = df)

# View summary
summary(model)

# Predict weight for new heights
new_heights <- data.frame(height = c(155, 165, 175))
predict(model, new_heights)

# Plot with regression line
plot(df$height, df$weight, pch=19, col="blue", main="Linear Regression Example",
     xlab="Height (cm)", ylab="Weight (kg)")
abline(model, col="red", lwd=2)
````

````r
# Load library
library(ggplot2)
library(gridExtra)

# -----------------------------
# 1. Positive Correlation
# -----------------------------
x1 <- c(1, 2, 3, 4, 5)
y1 <- c(2, 4, 5, 4, 6)
df1 <- data.frame(x=x1, y=y1)
corr1 <- cor(df1$x, df1$y)

p1 <- ggplot(df1, aes(x=x, y=y)) +
  geom_point(color="blue", size=3) +
  geom_smooth(method="lm", se=FALSE, color="red") +
  labs(title=paste("Positive Correlation\nr =", round(corr1,2)),
       x="X", y="Y") +
  theme_minimal()

# -----------------------------
# 2. Negative Correlation
# -----------------------------
x2 <- c(1, 2, 3, 4, 5)
y2 <- c(10, 8, 7, 5, 4)
df2 <- data.frame(x=x2, y=y2)
corr2 <- cor(df2$x, df2$y)

p2 <- ggplot(df2, aes(x=x, y=y)) +
  geom_point(color="green", size=3) +
  geom_smooth(method="lm", se=FALSE, color="red") +
  labs(title=paste("Negative Correlation\nr =", round(corr2,2)),
       x="X", y="Y") +
  theme_minimal()

# -----------------------------
# 3. No Correlation
# -----------------------------
set.seed(123)
x3 <- 1:10
y3 <- c(5,7,6,8,5,9,4,7,6,8)
df3 <- data.frame(x=x3, y=y3)
corr3 <- cor(df3$x, df3$y)

p3 <- ggplot(df3, aes(x=x, y=y)) +
  geom_point(color="purple", size=3) +
  geom_smooth(method="lm", se=FALSE, color="red") +
  labs(title=paste("No Correlation\nr =", round(corr3,2)),
       x="X", y="Y") +
  theme_minimal()

# -----------------------------
# 4. Linear Regression Example
# -----------------------------
height <- c(150, 160, 170, 180, 190)
weight <- c(50, 60, 65, 80, 90)
df4 <- data.frame(height=height, weight=weight)
model <- lm(weight ~ height, data=df4)
r2 <- summary(model)$r.squared

p4 <- ggplot(df4, aes(x=height, y=weight)) +
  geom_point(color="blue", size=3) +
  geom_smooth(method="lm", se=FALSE, color="red") +
  labs(title=paste("Linear Regression\nR-squared =", round(r2,2)),
       x="Height (cm)", y="Weight (kg)") +
  theme_minimal()

# -----------------------------
# Combine all plots
# -----------------------------
grid.arrange(p1, p2, p3, p4, ncol=2)
````
